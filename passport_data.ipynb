```python
import os
from datasets import Dataset

dataset = Dataset.from_file("C:\\Users\\nitis\\Documents\\Novac\\Doc_Class\\ind_passport\\ind_pass.arrow")

os.makedirs(
    "raw_dataset/passport",
    exist_ok=True
)

for i in range(len(dataset)):

    img = dataset[i]["image"]

    img.save(
        f"raw_dataset/passport/passport_{i}.jpg"
    )

print(
    f"Exported {len(dataset)} passport images"
)
```
